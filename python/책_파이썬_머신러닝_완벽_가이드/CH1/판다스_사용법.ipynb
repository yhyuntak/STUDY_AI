```python
# 기존의 index를 col으로 보내고 새롭게 인덱스 매기기

titanic_df = pd.read_csv('../dataset/titanic_train.csv')
titanic_reset_df = titanic_df.reset_index(inplace=False)
titanic_reset_df
```

```python
# 그러나 row를 slicing만 하는 것은 좋지 않다. loc,iloc을 애용하자
# iloc은 철저한 위치 기반 인덱싱만 허용한다.
print(titanic_df.head(3))
print()
print(titanic_df.iloc[0,0])
```

```python
# loc은 명칭 기반으로 row의 index 값, col의 feature 명을 작성해야함.
print(titanic_df.loc[0,'Survived'])
```

```python
# 당연히 둘다 슬라이싱이 되지만, 명칭 기반인 loc은 A:B 이면 A에서 B까지를 출력해주고, iloc은 A에서 B전까지 출력해준다.
```

```python
# 어떤 기준으로 정렬하기
print("바꾸기 전")
print(titanic_df.head(5))
titanic_sorted = titanic_df.sort_values(by=['Name'])
print("바꾼 후")
print(titanic_sorted.head(5))
```

```python
# 내림차순은 ascending=False 를 추가하자. 그리고 A로 정렬하고 A의 그룹 안에서 정렬은 다음과 같이
titanic_sorted = titanic_df.sort_values(by=['Pclass','Name'],ascending=False)
titanic_sorted.head(5)
```

```python
# aggregation을 column별로 가능
print(titanic_df[['Pclass','Age']].count())
print(titanic_df[['Pclass','Age']].max())
print(titanic_df[['Pclass','Age']].mean())
print(titanic_df[['Pclass','Age']].var())
```

```python
# 기준을 정하고 기준에 따라 그룹을 나누기
print(titanic_df.groupby(by='Pclass').count())
```

```python
# 데이터가 NaN인지 확인
print(titanic_df.isna().head(5))
print(titanic_df.isna().sum()) # NaN이 1로 처리되니, NaN의 개수를 세는 것
```

```python
# NaN 데이터 채우기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000') # inplace=False인 상태임.
titanic_df.head(4)
```

```python
# 그래서 fillna를 사용하면 결손 값을 평균으로 메꾸기도 가능하다.
titanic_df['Age'].fillna(titanic_df['Age'].mean(),inplace=True)
titanic_df.isna().sum()
```

```python
# lambda 식을 이용해서 col을 가져갈수도
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x)) # 네임에 있는 데이터들을 x로 불러와서 len(x) 처리하고 데이터로 삽입
titanic_df.loc[:titanic_df['Cabin'] = ,['Name','Name_len']]
```
