# 정확도(Accuracy)
---

정확도는 다음과 같은 계산으로 얻을 수 있다.

$$
정확도(Accuracy) = \frac{예측\;결과가\;정답과\;같은\;데이터\;수}{전체\;데이터\;수}
$$


그러나 이진 분류의 경우 데이터에 따라 정확도로 평가하는 것은 ML 모델을 정확히 평가하는 것이 아닐 수 있다.

그리고 만약 전체 데이터가 100개일 때, 클래스가 0 인 데이터가 1개, 클래스가 1인 데이터가 99개 있다면, 학습할 필요도 없이 그저 모두 클래스가 1이라고 판단하게되면 정확도가 99%가 된다. (극단적인 예지만..)

**따라서 정확도는 특히 불균형한 클래스 데이터 세트에서 성능을 판단할 때 사용하면 안된다.**

이런 한계점을 돌파하기 위해선 여러 가지 분류 지표를 함께 사용해야한다. 이제부터 한번 봐보자.

# 오차 행렬(confusion matrix)
---

오차 행렬은 분류 모델이 예측을 수행하면서 얼마나 헷갈리고 있는지 보여주는 지표이다.



```python
from IPython.display import Image
Image("images/오차행렬.png")
```

각각의 용어들에 대해선 다음을 참고하자. 다소 헷갈릴 수 있다.
뒤에 positive/negative는 예측한 결과를 뜻하고, 앞에 False/True는 정답과 비교해서 틀렸는가/맞췄는가를 의미한다.

* **False Positive(FP)** : <br/>
    예측 결과는 Positive인데 정답 클래스가 Negative라서 틀린 경우 (False는 예측과 결과가 다름을 뜻한다. Positive는 예측의 종류다.)
    
* **False Negative(FN)** : <br/>
    예측 결과는 Negative인데 정답 클래스가 Positive라서 틀린 경우 (False는 예측과 결과가 다름을 뜻한다. Negative는 예측의 종류다.)

* **True Positive(TP)** : <br/>
    예측 결과는 Positive인데 정답 클래스가 Positive라서 같은 경우 (True는 예측과 결과가 같음을 뜻한다. Positive는 예측의 종류다.)

* **True Negative(TN)** : <br/> 
    예측 결과는 Negative인데 정답 클래스가 Negative라서 같은 경우 (True는 예측과 결과가 같음을 뜻한다. Negative는 예측의 종류다.)
    

4개의 값은 정밀도(Precision), 재현율(Recall)값을 계산할 때 사용한다. 두 평가 지표는 정확도보다 더 선호된다.

<br/><br/>

# 정밀도
---

정밀도의 식은 다음과 같다.

$$
정밀도(precision) = \frac{TP}{FP+TP}
$$

말로 해석해보자면, positive로 예측한 결과들 중 정답과 일치하는 것들의 비율을 뜻한다. 즉, positive 예측 성능을 더욱 정밀하게 측정하기 위해서 사용된다. 정밀도가 작다는 것은, 예측은 positive인데 실제 값은 negative인 FP가 크다는 것이다. 즉, negative를 negative라고 꼭 예측해야하는 분야에서 사용되는 지표다. 예를 들면, 스팸 메일이 있다.

<br/><br/>

# 재현율
---

재현율의 식은 다음과 같다.

$$
재현율(Recall) = \frac{TP}{FN+TP}
$$

말로 해석해보면, 정답 레이블이 positive인 데이터들 중, 예측 값도 positive인 데이터의 비율을 뜻한다. 음.. 이 지표가 어디에 사용되는지 설명하기 위해서 예시를 보자. 

암이 의심되서 건강검진 받으러온 환자들 100명이 있다고 해보자. 환자들중 90명이 실제 암에 걸려있는 A그룹이고, 나머지 10명은 암에 걸리지 않은 B그룹이라고 하자. 이제 건강검진을 통해(머신러닝) 환자들의 상태를 결론냈다.(예측) 여기서 A그룹의 5명만, B그룹의 3명만 암에 걸렸다고 예측을 냈다고 하자. 그럼 TP는 5이다. FN은 85이다. 그럼 재현율은 식에 의해 $5/90=0.0555$ 이다. 

위 결과에서 보듯, 재현율이 낮으면 머신러닝 모델이 제대로된 예측을 하지 못해 최악의 결과를 냄을 볼 수 있다. 따라서 재현율은 positive를 positive로 꼭 예측해야하는 분야에서 사용된다. 

이제 예제로 살펴봐보자.

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder

def get_clf_eval(y_test,pred):
    confusion = confusion_matrix(y_test,pred)
    accuracy = accuracy_score(y_test,pred)
    precision = precision_score(y_test,pred)
    recall = recall_score(y_test,pred)
    
    print('오차 행렬')
    print(confusion)
    print("정확도 : {0:.4f}, 정밀도 : {1:.4f}, 재현율 : {2:.4f}".format(accuracy,precision,recall))


# Null 처리 함수
def fillna(df):
    df['Age'].fillna(df['Age'].mean(),inplace=True)
    df['Cabin'].fillna('N',inplace=True)
    df['Embarked'].fillna('N',inplace=True)
    df['Fare'].fillna(0,inplace=True)
    return df

# 머신러닝 알고리즘에 불필요한 속성 제거
def drop_features(df):
    df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
    return df

# 레이블 인코딩 수행. 
def format_features(df):
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin','Sex','Embarked']
    for feature in features:
        
        # labelencoder는 str형 데이터를 정수형 데이터로 바꿔주는 역할을 함.
        
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

# 앞에서 설정한 Data Preprocessing 함수 호출
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df



import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

titanic_df = pd.read_csv('../dataset/titanic_train.csv')
y_df = titanic_df['Survived']
X_df = titanic_df.drop(['Survived'],axis=1)
X_df = transform_features(X_df) # Nan값, 필요없는 col제거, str형 데이터를 정수형으로 바꾼 결과
X_df.head(5)
```

```python
# 데이터 분리
X_train,X_test,y_train,y_test = train_test_split(X_df,y_df,test_size=0.2,random_state=1)

# 모델 생성
lg_clf = LogisticRegression()
# 학습
lg_clf.fit(X_train,y_train)
# 예측
pred = lg_clf.predict(X_test)

get_clf_eval(y_test,pred)
```

정밀도와 재현율은 둘다 같이 1에 가까운 수를 얻을 수록 좋은 모델을 얻었다고 평가할 수 있다. 그러나 위 결과는 그렇게까지 되진 않았고 심지어 재현율은 더 떨어진다. 

# 정밀도와 재현율의 trade-off
---

둘다 올리면 좋지만.. 한쪽을 올리면 다른 한쪽은 내려가게 된다. 이를 trade-off 관계라고 부른다. 

사이킷런은 분류 결정 threshold를 조절하여 trade-off를 조절하는데, 분류 결정 threshold란, 모델로부터 입력 데이터의 클래스 각각에 대한 사후 확률을 얻게되었을 때, 어떤 클래스로 분류할지 기준으로 사용되는 값이다. 간단한 예시는 다음과 같다.

```python
from sklearn.preprocessing import Binarizer

X = [[1,-1,3],[3,-1,-2],[-2.1,1.1,3.1]]
binarizer = Binarizer(threshold=1.1)
print(binarizer.fit_transform(X))
```

실제로 사이킷런은 Binarizer를 사용해서 클래스를 분류한다고 한다.

이제 이 threshold 값에 따라 정밀도와 재현율이 어떻게 바뀌는지 봐보자.


## threshold = 0.5

```python
from sklearn.preprocessing import Binarizer

#분류 결정 임곗값
custom_threshold = 0.5

pred_proba = lg_clf.predict_proba(X_test)

# predict_proba( ) 반환값의 두번째 컬럼 , 즉 Positive 클래스 컬럼 하나만 추출하여 Binarizer를 적용
pred_proba_1 = pred_proba[:,1].reshape(-1,1)

binarizer = Binarizer(threshold=custom_threshold).fit(pred_proba_1) 
custom_predict = binarizer.transform(pred_proba_1)

get_clf_eval(y_test, custom_predict)
```

## threshold = 0.4

```python
custom_threshold = 0.4
pred_proba_1 = pred_proba[:,1].reshape(-1,1)
binarizer = Binarizer(threshold=custom_threshold).fit(pred_proba_1) 
custom_predict = binarizer.transform(pred_proba_1)

get_clf_eval(y_test , custom_predict)
```

임계값을 낮췄더니 재현율이 올라가고 정밀도가 떨어졌다. 그 이유는 TP의 값이 커지고 FN의 값이 작아지기 때문이다. 다시 말하자면, 임계값이 작아졌으니 positve라고 예측하는 비율이 늘어났을 것인데.. 그렇다는 것은 정답 클래스가 positive로 예측과 맞는 개수 늘어나 TP가 더 커지게 될 것이고, Negative로 예측하는 비율이 줄어서 FN의 값이 작아지기 때문이다. 

그러나 정밀도가 낮아지게 되었다. 그 이유는 Positive라고 예측한 비율이 늘으면서, 실제 클래스가 Negative인 것도 Positive라고 보는 데이터들이 많아지게 되어서 FP가 늘어났기 때문이다.

그럼 임계값을 조금씩 늘려서 확인해 볼까? 사이킷런은 이를 위한 API가 존재한다. 아래 예시를 보자.

```python
from sklearn.metrics import precision_recall_curve
import numpy as np

pred_proba_class1 = lg_clf.predict_proba(X_test)[:, 1] # positve의 확률만 가져가기.

# 실제값 데이터 셋과 레이블 값이 1일 때의 예측 확률을 precision_recall_curve 인자로 입력 
precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_class1 )
print('반환된 분류 결정 임곗값 배열의 Shape:', thresholds.shape)
print('반환된 precisions 배열의 Shape:', precisions.shape)
print('반환된 recalls 배열의 Shape:', recalls.shape)

print("thresholds 5 sample:", thresholds[:5])
print("precisions 5 sample:", precisions[:5])
print("recalls 5 sample:", recalls[:5])

#반환된 임계값 배열 로우가 163건이므로 샘플로 10건만 추출하되, 임곗값을 16 Step으로 추출. 
thr_index = np.arange(0, thresholds.shape[0], 16)
print('샘플 추출을 위한 임계값 배열의 index 10개:', thr_index)
print('샘플용 10개의 임곗값: ', np.round(thresholds[thr_index], 2))

# 16 step 단위로 추출된 임계값에 따른 정밀도와 재현율 값 
print('샘플 임계값별 정밀도: ', np.round(precisions[thr_index], 3))
print('샘플 임계값별 재현율: ', np.round(recalls[thr_index], 3))
```

위 결과를 통해 임계값이 조절됨에따라 정밀더와 재현율이 바뀌는 것을 알 수 있다!

precision_recall_curve() API는 임계값에 따른 정밀도와 재현율의 값 변화 그래프를 그래프로 시각화하는데 사용할 수 있다.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

def precision_recall_curve_plot(y_test , pred_proba_c1):
    # threshold ndarray와 이 threshold에 따른 정밀도, 재현율 ndarray 추출. 
    precisions, recalls, thresholds = precision_recall_curve( y_test, pred_proba_c1)
    
    # X축을 threshold값으로, Y축은 정밀도, 재현율 값으로 각각 Plot 수행. 정밀도는 점선으로 표시
    plt.figure(figsize=(8,6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds,precisions[0:threshold_boundary],linestyle='--',label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
    
    # threshold 값 X 축의 Scale을 0.1 단위로 변경
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    
    # x축, y축 label과 legend, 그리고 grid 설정
    plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
    plt.legend(); plt.grid()
    plt.show()
    
precision_recall_curve_plot( y_test, lg_clf.predict_proba(X_test)[:, 1] )
```

가장 좋은 분류 모델의 결과를 찾는 방법은 두 지표가 둘다 높을 때이다. 아마 위 그래프에서 교차 지점을 뜻하지 않을까?

# F1 스코어
---

F1 스코어는 정밀더와 재현율을 결합한 지표이다. F1 스코어는 정밀도와 재현율이 어느 한쪽으로 치우치지 않을 때, 상대적으로 높은 값을 가진다. 식은 다음과 같다.

$$
F1 = \frac{2}{\frac{1}{recall}+\frac{1}{precision}} = 2*\frac{precision * recall}{precision + recall}
$$

예제를 봐보자.

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test,pred)
print("F1 스코어 : {0:.4f}".format(f1))
```

# ROC 곡선
---

ROC(receiver operation characteristic) 곡선은 FPR(False Positive Rate)이 변할 때, TPR(True Positive Rate)이 어떻게 변하는지 나타내는 곡선이다. FPR을 x축, TPR을 y축으로 잡는다. 

TPR은 재현율을 나타낸다. 다른 말로 민감도라고도 불린다. 반대로 TNR(True Negative Rate)이라고 불리는 특이성(Specificity)가 있는데, TPR은 실제 positive가 정확히 예측되어야하는 수준을, TNR은 실제 negative가 정확히 예측되는 수준을 나타낸다. 

TNR은 TN/(FP+TN)의 식으로 구해진다. FPR은 FP/(FP+TN)으로 계산되며, 1-TNR이라고도 볼 수 있다. 

이제 ROC 곡선의 예를 봐보자. 

```python
from sklearn.metrics import roc_curve

# 레이블 값이 1일때의 예측 확률을 추출 
pred_proba_class1 = lg_clf.predict_proba(X_test)[:,1]

FPRs , TPRs , thresholds = roc_curve(y_test, pred_proba_class1)
# 반환된 임곗값 배열에서 샘플로 데이터를 추출하되, 임곗값을 8 Step으로 추출. 
# thresholds[0]은 max(예측확률)+1로 임의 설정됨. 이를 제외하기 위해 np.arange는 1부터 시작
thr_index = np.arange(1,thresholds.shape[0],8)
print('샘플 추출을 위한 임곗값 배열의 index:', thr_index)
print('샘플 index로 추출한 임곗값: ', np.round(thresholds[thr_index], 2))

# 5 step 단위로 추출된 임계값에 따른 FPR, TPR 값
print('샘플 임곗값별 FPR: ', np.round(fprs[thr_index], 3))
print('샘플 임곗값별 TPR: ', np.round(tprs[thr_index], 3))
```

```python
def roc_curve_plot(y_test , pred_proba_c1):
    # 임곗값에 따른 FPR, TPR 값을 반환 받음. 
    FPRs,TPRs,thresholds = roc_curve(y_test,pred_proba_c1)
    
    # ROC Curve를 plot 곡선으로 그림. 
    plt.plot(FPRs,TPRs,label='ROC')
    # 가운데 대각선 직선을 그림. 
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    
    # FPR X 축의 Scale을 0.1 단위로 변경, X,Y 축명 설정등   
    start, end = plt.xlim() # 현재 그래프의 x축의 limit을 호출
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    plt.xlim(0,1); plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )'); plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()
    
roc_curve_plot(y_test, lg_clf.predict_proba(X_test)[:, 1] )
```

ROC 곡선이 가운데 그어져있는 대쉬 라인의 직선에서 멀어질수록 성능이 뛰어남을 의미한다. 

근데 ROC 곡선은 그저 TPR과 FPR의 관계를 눈으로 볼 수 있을 뿐이다. 수치적으로 평가 지표를 낼 수 없다. 
따라서 분류의 성능 지표로썬 ROC 곡선 면적에 기반한 AUC 값으로 결정한다.

# AUC
---

AUC(area under curve)는 ROC 곡선의 면적을 구한 것으로, 일반적으로 1에 가까울수록 좋은 수치를 뜻한다. 즉, ROC곡선이 직선과 완전 멀어져서 
[FPR,TPR]=[0,1] 에 닿으면 면적이 1이 되기 때문이다. 

위 그래프의 AUC를 구하면 다음과 같다.

```python
from sklearn.metrics import roc_auc_score

### 아래는 roc_auc_score()의 인자를 잘못 입력한 것으로, 책에서 수정이 필요한 부분입니다. 
### 책에서는 roc_auc_score(y_test, pred)로 예측 타겟값을 입력하였으나 
### roc_auc_score(y_test, y_score)로 y_score는 predict_proba()로 호출된 예측 확률 ndarray중 Positive 열에 해당하는 ndarray입니다. 

#pred = lr_clf.predict(X_test)
#roc_score = roc_auc_score(y_test, pred)

pred_proba = lg_clf.predict_proba(X_test)[:, 1]
roc_score = roc_auc_score(y_test, pred_proba)
print('ROC AUC 값: {0:.4f}'.format(roc_score))
```
